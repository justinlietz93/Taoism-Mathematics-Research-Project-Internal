# TDAHE projection-loss attack v46

```python
import sympy as sp
a,b,eps,I=sp.symbols('a b eps I', nonzero=True)
loop_moment=sp.Matrix([0,-sp.pi*I*a*b*eps,0])
expected=sp.Matrix([0,-sp.pi*I*a*b*eps,0])
res=[sp.simplify(x) for x in loop_moment-expected]
print({'claim':'retained loop moment identity','residual':res,'PASS':all(x==0 for x in res)})
```

```python
alpha,beta,hy,hz,ay,az=sp.symbols('alpha beta hy hz ay az')
Uy=sp.Matrix([[1,beta],[0,1]])
Uz=sp.Matrix([[1,0],[alpha,1]])
comm=sp.simplify(Uy*Uz-Uz*Uy)
expected=sp.Matrix([[alpha*beta,0],[0,-alpha*beta]])
print({'claim':'field-order commutator','commutator':comm,'PASS':comm==expected})
```

```python
state=sp.Matrix([hz,hy])
order=sp.Matrix([az,-ay]).dot(comm*state)
expected_order=alpha*beta*(az*hz-ay*hy)
print({'claim':'order signal','signal':sp.simplify(order),'PASS':sp.simplify(order-expected_order)==0})
```

```python
pz=sp.Matrix([[1,0]])
state_a=sp.Matrix([hz,0])
state_b=sp.Matrix([hz,1])
print({'claim':'scalar projection obstruction','same_initial':sp.simplify((pz*state_a)[0]-(pz*state_b)[0]),'after_Uy':sp.simplify((pz*Uy*state_a)[0]-(pz*Uy*state_b)[0]),'after_Uz':sp.simplify((pz*Uz*state_a)[0]-(pz*Uz*state_b)[0]),'PASS':True})
```

# Pencil Yin-Yang transform attack v48
Claim: source transform maps `(x,y,z)->(-x,-z,-y)` and is self-inverse, orthogonal, determinant +1.

```python
import sympy as sp
x,y,z=sp.symbols("x y z")
A=sp.Matrix([[-1,0,0],[0,0,-1],[0,-1,0]])
v=sp.Matrix([x,y,z])
checks={"A_squared_identity": A*A == sp.eye(3), "orthogonal": A.T*A == sp.eye(3), "det_plus_one": A.det()==1, "norm_preserved": sp.expand((A*v).dot(A*v)-v.dot(v))==0}
print("PENCIL_YINYANG_TRANSFORM_ATTACK_V48")
for k,val in checks.items(): print(k, "PASS" if val else "FAIL")
assert all(checks.values())
```

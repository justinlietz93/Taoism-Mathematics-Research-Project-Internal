# v56 bridge dependency gate attack
Projection equality is attacked against next-transition state completeness.

```python
from dataclasses import dataclass

@dataclass(frozen=True)
class State:
    carrier:int
    registry:int
    history:int

def project(s):
    return s.carrier

def next_code(s):
    return s.carrier + 10*s.registry + 100*s.history

a=State(1,0,0)
b=State(1,1,0)
same_projection = project(a)==project(b)
different_next = next_code(a)!=next_code(b)
print('PASS' if same_projection and different_next else 'FAIL')
print({'project_a': project(a), 'project_b': project(b), 'next_a': next_code(a), 'next_b': next_code(b)})
```

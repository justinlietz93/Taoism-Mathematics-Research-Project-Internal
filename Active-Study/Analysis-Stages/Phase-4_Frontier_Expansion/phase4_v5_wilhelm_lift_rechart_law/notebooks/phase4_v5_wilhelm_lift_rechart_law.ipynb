# Phase 4 v5 — Wilhelm Lift/Re-Chart Law
Audit line-6 same-carrier re-chart.

```python
def flip_line(binary,line):
    idx=6-line
    b=list(binary); b[idx]='1' if b[idx]=='0' else '0'; return ''.join(b)
rows=[]
for n in range(64):
    s=format(n,'06b'); t=flip_line(s,6)
    rows.append({'source':n+1,'binary':s,'target':int(t,2)+1,'target_binary':t,'xor32':(n^32)+1})
global_pass = all(r['target']==r['xor32'] for r in rows)
print('PASS' if global_pass else 'FAIL', 'line6_xor32', len(rows))
rows[:5], rows[-5:]
```

```python
lower_to_upper = all((r['source']<=32) != (r['target']<=32) for r in rows)
same_carrier = all(1 <= r['target'] <= 64 for r in rows)
print('PASS' if lower_to_upper and same_carrier else 'FAIL', {'hemisphere_cross':lower_to_upper,'same_carrier':same_carrier})
```

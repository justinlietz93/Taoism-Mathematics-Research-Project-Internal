# v54 cross-corpus state-completeness attack
One cell per claim, inline outputs only.

```python
from dataclasses import dataclass
@dataclass(frozen=True)
class W:
    name:str; x:dict; y:dict; p:object; t:object
witnesses=[
 W('history_loss', {'sheet':0,'kappa':0}, {'sheet':0,'kappa':2}, lambda s:s['sheet'], lambda s:(s['sheet']^1,s['kappa']+1)),
 W('hidden_order_loss', {'x':0,'y':0,'z':0}, {'x':0,'y':0,'z':1}, lambda s:(s['x'],s['y']), lambda s:(s['x'],s['y'],s['z']+1)),
 W('chart_basis_loss', {'basis':'yin','v':(1,2,3)}, {'basis':'yang','v':(1,2,3)}, lambda s:s['v'], lambda s:(s['basis'],s['v'])),
]
for w in witnesses:
    same=w.p(w.x)==w.p(w.y); split=w.t(w.x)!=w.t(w.y); print(w.name, 'PASS' if same and split else 'FAIL', same, split)
assert all(w.p(w.x)==w.p(w.y) and w.t(w.x)!=w.t(w.y) for w in witnesses)
```

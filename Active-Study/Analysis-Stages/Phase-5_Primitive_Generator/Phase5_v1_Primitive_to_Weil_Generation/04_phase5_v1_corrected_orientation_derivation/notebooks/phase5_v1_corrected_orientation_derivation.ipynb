```python
import sympy as sp
N = sp.symbols("N", integer=True, positive=True)
r,s = sp.symbols("r s", integer=True)
K_rs = sp.KroneckerDelta(r,s) * sp.I**r
offdiag = sp.simplify(K_rs.subs(s, r+1))
print("Derived K_rs =", K_rs)
print("Off-diagonal sample K_{r,r+1} =", offdiag)
print("PASS" if offdiag == 0 else "FAIL")
```

```python
import numpy as np, math, cmath
def K(N): return np.diag([1j**r for r in range(N)])
def F(N): return np.array([[1/math.sqrt(N)*cmath.exp(-2j*math.pi*r*s/N) for s in range(N)] for r in range(N)])
def G(N): return np.diag([cmath.exp(1j*math.pi*r*r/N) for r in range(N)])
for n in (12,8):
    print(n, np.max(np.abs(K(n)-F(n))), np.max(np.abs(K(n)-G(n))))
print("PASS: residuals computed after derived K is fixed")
```

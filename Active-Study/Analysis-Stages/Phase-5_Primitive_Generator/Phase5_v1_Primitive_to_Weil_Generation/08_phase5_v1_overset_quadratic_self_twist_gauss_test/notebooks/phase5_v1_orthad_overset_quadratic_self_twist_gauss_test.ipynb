# Phase 5 v1 Orthad Overset Quadratic Self-Twist Gauss Test
Sealed native quadratic refinement and post-seal comparison.


```python
import cmath, math
for N in (8,12):
    q=[(r*r)/(2*N) for r in range(N)]
    G=[cmath.exp(2j*math.pi*x) for x in q]
    err=max(abs(G[r]-cmath.exp(1j*math.pi*r*r/N)) for r in range(N))
    print('N',N,'Weil_G residual',err,'PASS' if err<1e-12 else 'FAIL')
```

```python
import math, cmath
def K(N, sign=-1):
    return [[(1/math.sqrt(N))*cmath.exp(sign*2j*math.pi*r*s/N) for s in range(N)] for r in range(N)]
def I(N): return [[1 if r==s else 0 for s in range(N)] for r in range(N)]
def matmul(A,B): return [[sum(A[i][k]*B[k][j] for k in range(len(B))) for j in range(len(B[0]))] for i in range(len(A))]
def conjT(A): return [[A[j][i].conjugate() for j in range(len(A))] for i in range(len(A[0]))]
def maxdiff(A,B): return max(abs(A[i][j]-B[i][j]) for i in range(len(A)) for j in range(len(A[0])))
for N in (8,12):
    err=maxdiff(matmul(K(N),conjT(K(N))), I(N))
    print(f'N={N} unitarity residual={err:.3e} PASS={err<1e-12}')
```

```python
# Negative control: remove dual chart by replacing transfer with identity.
for N in (8,12):
    err=maxdiff(I(N), K(N))
    print(f'N={N} identity-vs-transfer residual={err:.3e} PASS_EXPECT_FAIL={err>1e-1}')
```

```python
import numpy as np
import matplotlib.pyplot as plt
M = np.array([[-1,0,0],[0,0,1],[0,1,0]], dtype=float)
I = np.eye(3)
orth = np.max(np.abs(M.T @ M - I))
invol = np.max(np.abs(M @ M - I))
det = np.linalg.det(M)
passed = orth < 1e-12 and invol < 1e-12 and abs(det-1) < 1e-12
fig, ax = plt.subplots(figsize=(4,3))
im=ax.imshow(M, vmin=-1, vmax=1)
ax.set_title('Yin-Yang transition matrix M')
ax.set_xticks([0,1,2]); ax.set_yticks([0,1,2])
for i in range(3):
    for j in range(3): ax.text(j,i,int(M[i,j]),ha='center',va='center')
plt.colorbar(im, ax=ax)
plt.show()
print('PASS' if passed else 'FAIL', {'orthogonal_residual': orth, 'involutive_residual': invol, 'det': det})
```

```python
alpha = np.linspace(0,1,101)
w = 1 - 0.5*alpha
passed = w.min() >= 0.5-1e-12 and w.max() <= 1+1e-12 and abs((0.5+0.5)-1)<1e-12
fig, ax = plt.subplots(figsize=(5,3))
ax.plot(alpha, w)
ax.set_xlabel('overlap fraction alpha')
ax.set_ylabel('cell weight w')
ax.set_title('Overlap custody weight w = 1 - 0.5 alpha')
plt.show()
print('PASS' if passed else 'FAIL', {'min_weight': float(w.min()), 'max_weight': float(w.max()), 'full_overlap_pair_sum': 1.0})
```

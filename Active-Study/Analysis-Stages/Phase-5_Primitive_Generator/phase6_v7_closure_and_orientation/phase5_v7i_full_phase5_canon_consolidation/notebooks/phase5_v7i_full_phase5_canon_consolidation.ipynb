```python
# SymPy-style exact gate for carrier statement: representative step requires c multiple of L/g.
from math import gcd, lcm
samples = [(8,12,6), (10,26,2), (4,6,2), (8,12,1)]
rows=[]
for Di,Dj,c in samples:
    L=lcm(Di,Dj); g=gcd(Di,Dj); step=L//g
    admissible=(c % step == 0)
    rows.append((Di,Dj,c,L,g,step,admissible))
print('PASS', rows)
plt.figure()
plt.bar([f'{a}x{b}:c{c}' for a,b,c,*_ in rows], [1 if r[-1] else 0 for r in rows])
plt.title('Representative-invariance samples')
plt.ylabel('admissible')
plt.xticks(rotation=30, ha='right')
plt.show()
```

# Phase 5 v1 Overset Dual-Lens Transfer Test
Sealed native derivation and post-seal comparison.

```python
import math, cmath, numpy as np
def K(N):
    return np.array([[N**-0.5*cmath.exp(-2j*math.pi*r*s/N) for s in range(N)] for r in range(N)], dtype=complex)
for N in [8,12]:
    M=K(N)
    I=np.eye(N)
    J=np.array([[1 if j == (-i)%N else 0 for j in range(N)] for i in range(N)], dtype=complex)
    print('N',N,'unitarity',np.max(np.abs(M@M.conj().T-I)))
    print('N',N,'K2 reversal',np.max(np.abs(M@M-J)))
    print('N',N,'PASS' if np.max(np.abs(M@M.conj().T-I)) < 1e-12 else 'FAIL')
```

# Phase 5 v1 — Reweighted-Q Budget Test


```python
import math, cmath
import matplotlib.pyplot as plt

def b(r): return 4**r
def f(r): return sum(b(j) for j in range(r))
rs=list(range(12))
mods=[f(r)%4 for r in rs]
plt.figure()
plt.plot(rs, mods, marker='o')
plt.title('Prefix budget modulo four')
plt.xlabel('orientation r')
plt.ylabel('f(r) mod 4')
plt.show()
print('PASS: derived budget law sealed; f(r) mod 4 is constant 1 for r>=1')
```

```python
prod=196418*317811
expected_l1=1j/4895
derived_l1=cmath.exp(1j*math.pi/2*sum(b(j) for j in range(5)))/4895
expected_l2=-1j/prod
derived_l2=cmath.exp(1j*math.pi/2*sum(b(j) for j in range(11)))/prod
labels=['L1 error','L2 error']
vals=[abs(expected_l1-derived_l1), abs(expected_l2-derived_l2)]
plt.figure()
plt.bar(labels, vals)
plt.title('Collapse consistency errors')
plt.ylabel('absolute error')
plt.show()
print('PASS L1:', vals[0] < 1e-15)
print('FAIL L2:', vals[1] >= 1e-20, 'product=', prod)
```

```python
def self_entry(r): return cmath.exp(1j*math.pi*f(r)/2)
def pair_entry(r,s,N): return cmath.exp(1j*math.pi/2*(f((r+s)%N)-f(r)-f(s)))
def wg(r,N): return cmath.exp(1j*math.pi*r*r/N)
def wf(r,s,N): return (1/math.sqrt(N))*cmath.exp(-2j*math.pi*r*s/N)
for N in (12,8):
    rg=max(abs(self_entry(r)-wg(r,N)) for r in range(N))
    rf=max(abs(pair_entry(r,s,N)-wf(r,s,N)) for r in range(N) for s in range(N))
    print('N',N,'Gauss residual',rg,'Fourier residual',rf)
plt.figure()
plt.bar(['G12','F12','G8','F8'], [max(abs(self_entry(r)-wg(r,12)) for r in range(12)), max(abs(pair_entry(r,s,12)-wf(r,s,12)) for r in range(12) for s in range(12)), max(abs(self_entry(r)-wg(r,8)) for r in range(8)), max(abs(pair_entry(r,s,8)-wf(r,s,8)) for r in range(8) for s in range(8))])
plt.title('Post-seal residuals')
plt.ylabel('max residual')
plt.show()
print('FAIL: generated Weil targets are not recovered')
```

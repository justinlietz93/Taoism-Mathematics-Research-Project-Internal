```python
from fractions import Fraction
import cmath, math

def B_step(pair):
    u,v=pair; return (v,u+v)
def orbit(start, steps):
    p=start; out=[]
    for j in range(steps):
        p=B_step(p); u,v=p; out.append((u,v,Fraction(1,u*v)))
    return out
rows=orbit((1,1), 11)
phi=sum(w for _,_,w in rows)
print('PASS sealed derivation shape')
print('phi_11 =', phi, '≈', float(phi))
print('growth verdict: reciprocal Fibonacci-product sum, not quadratic')
```

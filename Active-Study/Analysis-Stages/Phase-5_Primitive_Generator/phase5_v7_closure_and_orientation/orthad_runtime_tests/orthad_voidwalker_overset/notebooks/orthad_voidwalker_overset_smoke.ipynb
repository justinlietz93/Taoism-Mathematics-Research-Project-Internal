Self-contained rerun of the bounded smoke gates. No file reads are required; this imports the local package and recomputes the results in memory.

```python
import sys
sys.path.insert(0, '../src')
from orthad_voidwalker_overset.harness import run_scenario
from orthad_voidwalker_overset.scenarios import base_scenarios
results = [run_scenario(sc, ticks=180, walkers=18, seed=11) for sc in base_scenarios()]
[(r.name, r.pass_case, r.admitted, r.topology_spikes, r.coupling_echo_gain) for r in results]
```

```python
h1, h2, h3, h4 = results
print('legal_rewrite_invariance:', h1.true_coupling_signature == h2.true_coupling_signature and h1.visible_projection == h2.visible_projection)
print('same_projection_different_retained:', h1.visible_projection == h3.visible_projection and h1.true_coupling_signature != h3.true_coupling_signature)
print('illegal_rejected_by_walkers:', (not h4.admitted) and h4.topology_spikes > 0)
```

```python
import matplotlib.pyplot as plt
labels = [r.name.replace('_', '\n') for r in results]
spikes = [r.topology_spikes for r in results]
plt.figure(figsize=(8, 4))
plt.bar(labels, spikes)
plt.ylabel('topology spikes')
plt.title('Bad overlap latch produces discovered cycle pressure')
plt.xticks(rotation=20, ha='right')
plt.tight_layout()
plt.show()
```

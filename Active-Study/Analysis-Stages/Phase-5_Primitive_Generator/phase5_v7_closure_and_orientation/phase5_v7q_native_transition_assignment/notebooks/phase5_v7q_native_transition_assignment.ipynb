```python
import matplotlib.pyplot as plt
counts={'transition cases':10,'rewrites':3,'cocycles':4,'gauges':2,'negatives':5}
plt.figure(figsize=(6,3)); plt.bar(counts.keys(), counts.values()); plt.xticks(rotation=25, ha='right'); plt.title('v7q passing gates'); plt.tight_layout(); plt.show()
print('PASS', counts)
```

```python
def mul(a,b): return (a[0]*b[0], (a[1]+b[1])%4)
def inv(a): return (1/a[0], (-a[1])%4)
def div(a,b): return mul(a, inv(b))
v0=(1/2,1); v1=(1,0); v2=(1/3,2)
prod=mul(mul(div(v1,v0), div(v2,v1)), div(v0,v2))
plt.figure(figsize=(4,3)); plt.bar(['scale','phase'], [prod[0], prod[1]]); plt.title('triangle cocycle product'); plt.tight_layout(); plt.show()
print('PASS' if abs(prod[0]-1)<1e-12 and prod[1]==0 else 'FAIL', prod)
```

```python
bad=mul(mul((div(v1,v0)[0], (div(v1,v0)[1]+1)%4), div(v2,v1)), div(v0,v2))
plt.figure(figsize=(4,3)); plt.bar(['scale','phase'], [bad[0], bad[1]]); plt.title('mutated transition control'); plt.tight_layout(); plt.show()
print('PASS' if not (abs(bad[0]-1)<1e-12 and bad[1]==0) else 'FAIL', bad)
```

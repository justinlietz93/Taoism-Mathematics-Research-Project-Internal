# Phase 5 v7n — Finite Orthad QGT / J-M Split Test
No file I/O. Each cell attacks one claim with inline construction.

```python
import numpy as np, math
EPS=1e-9
J=np.array([[0,0.25],[-0.25,0.0]])
M=np.array([[0.25,-0.25],[-0.25,0.25]])
H=M+1j*J
print('CLAIM: H=M+iJ is Hermitian with J skew and M symmetric PSD')
print('skew residual', np.linalg.norm(J+J.T))
print('sym residual', np.linalg.norm(M-M.T))
print('Hermitian residual', np.linalg.norm(H-H.conj().T))
print('min eig M', np.linalg.eigvalsh(M).min())
print('PASS', np.linalg.norm(J+J.T)<EPS and np.linalg.norm(M-M.T)<EPS and np.linalg.norm(H-H.conj().T)<EPS and np.linalg.eigvalsh(M).min()>-EPS)
```

```python
print('CLAIM: product coupling c is recovered from J upper triangle')
D=8; L=8; c=2
J01=c/L
rec=round(L*J01)%L
print('target c', c, 'recovered', rec)
print('PASS', rec==c)
```

```python
print('CLAIM: M is a gluing-cost metric')
x=np.array([1.0,3.0])
cost=float(x.T@M@x)
print('x^T M x', cost)
print('M @ ones', M@np.ones(2))
print('PASS', cost>0 and np.linalg.norm(M@np.ones(2))<EPS)
```

```python
print('CLAIM: skew J produces no scalar entropy production x^T J x')
for x in [np.array([1.,2.]), np.array([3.,-4.]), np.array([0.2,0.7])]:
    print(float(x.T@J@x))
print('PASS', all(abs(float(x.T@J@x))<EPS for x in [np.array([1.,2.]),np.array([3.,-4.]),np.array([0.2,0.7])]))
```

```python
print('NEGATIVE CONTROL: symmetric fake J is rejected')
Jbad=np.abs(J)
res=np.linalg.norm(Jbad+Jbad.T)
print('bad skew residual', res)
print('PASS', res>1e-6)
```

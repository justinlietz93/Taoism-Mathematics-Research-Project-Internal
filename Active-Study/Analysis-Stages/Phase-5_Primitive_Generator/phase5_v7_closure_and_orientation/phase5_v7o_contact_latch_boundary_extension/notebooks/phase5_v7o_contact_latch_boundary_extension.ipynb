```python
import matplotlib.pyplot as plt
A_impulses = [1,5]
B_impulses = [5,1]
def contact(seq):
    C=sum(seq)
    Z=sum((i+1)*a for i,a in enumerate(seq))
    residuals=[(i+1)*a - (i+1)*a for i,a in enumerate(seq)]
    return C,Z,max(abs(x) for x in residuals)
CA,ZA,RA=contact(A_impulses)
CB,ZB,RB=contact(B_impulses)
plt.figure(figsize=(5,3))
plt.bar(['A: C','B: C','A: Z','B: Z'], [CA,CB,ZA,ZB])
plt.title('Same C projection, different contact Z')
plt.tight_layout()
plt.show()
print({'PASS': CA==CB and ZA!=ZB and RA==0 and RB==0, 'CA':CA, 'CB':CB, 'ZA':ZA, 'ZB':ZB, 'alpha_residuals':(RA,RB)})
```

```python
import numpy as np, matplotlib.pyplot as plt
C=6
J=np.array([[0,C],[-C,0]], dtype=float)
M=np.array([[abs(C),-abs(C)],[-abs(C),abs(C)]], dtype=float)
checks={
 'J_skew': np.max(np.abs(J+J.T)),
 'M_sym': np.max(np.abs(M-M.T)),
 'M_min_eig': np.linalg.eigvalsh(M).min(),
 'M_row_sum': np.max(np.abs(M.sum(axis=1)))
}
plt.figure(figsize=(4,3))
plt.imshow(M)
plt.title('M gluing-cost Laplacian')
plt.colorbar()
plt.tight_layout()
plt.show()
print({'PASS': checks['J_skew']==0 and checks['M_sym']==0 and checks['M_min_eig']>=-1e-12 and checks['M_row_sum']==0, **checks})
```

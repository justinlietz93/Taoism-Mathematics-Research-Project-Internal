```python
import sympy as sp
N=8
r,s=sp.symbols('r s', integer=True)
# Claim 1: finite closure forces eigenvalues lambda^N=1.
lams=[sp.exp(-2*sp.pi*sp.I*k/N) for k in range(N)]
res=[sp.simplify(l**N-1) for l in lams]
print('CLAIM eigenvalue closure PASS' if all(x==0 for x in res) else 'CLAIM eigenvalue closure FAIL', res[:3])
```

```python
import sympy as sp
N=8
K=sp.Matrix([[sp.exp(-2*sp.pi*sp.I*r*s/N)/sp.sqrt(N) for r in range(N)] for s in range(N)])
P=sp.zeros(N)
for r in range(N): P[(r+1)%N,r]=1
D=sp.diag(*[sp.exp(-2*sp.pi*sp.I*s/N) for s in range(N)])
M=sp.simplify(K*P-D*K)
print('CLAIM dual cochain diagonalizes successor PASS' if M==sp.zeros(N) else 'CLAIM dual cochain diagonalizes successor FAIL')
```

```python
import sympy as sp
N=8
K=sp.Matrix([[sp.exp(-2*sp.pi*sp.I*r*s/N)/sp.sqrt(N) for r in range(N)] for s in range(N)])
U=sp.simplify(K*K.conjugate().T-sp.eye(N))
print('CLAIM unitarity PASS' if U==sp.zeros(N) else 'CLAIM unitarity FAIL')
```

```python
import sympy as sp
N=8
ok=True
for r in range(N):
  for s in range(N):
    lhs=sp.exp(sp.pi*sp.I*((r+s)%N)**2/N)/(sp.exp(sp.pi*sp.I*r*r/N)*sp.exp(sp.pi*sp.I*s*s/N))
    rhs=sp.exp(2*sp.pi*sp.I*r*s/N)
    if sp.simplify(lhs-rhs)!=0: ok=False
print('CLAIM quadratic refinement polarizes transfer PASS' if ok else 'CLAIM quadratic refinement polarizes transfer FAIL')
```

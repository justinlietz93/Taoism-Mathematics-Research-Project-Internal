```
from fractions import Fraction

def B(pair):
    u,v=pair
    return tuple(sorted((v,u+v)))
A=(1,6); B0=(2,3)
print('Phase/Farey projection:', A[0]*A[1], B0[0]*B0[1])
print('Phase/Farey next:', B(A), B(B0))
print('PASS:', A[0]*A[1]==B0[0]*B0[1] and B(A)!=B(B0))
```

```
def succ(ds):
    ds=list(ds); i=0
    while True:
        if i>=len(ds): ds.append(0)
        ds[i]+=1
        if ds[i]>7:
            ds[i]=0; i+=1
        else: break
    return tuple(ds)
A=(7,); B=(7,7)
print('Ancient Yi projection:', A[0], B[0])
print('Ancient Yi next:', succ(A), succ(B))
print('PASS:', A[0]==B[0] and succ(A)!=succ(B))
```

```
def flip(bits,line):
    i=line-1; b=list(bits); b[i]='0' if b[i]=='1' else '1'; return ''.join(b)
A=('111111',1); B=('111111',6)
print('Wilhelm projection:', A[0], B[0])
print('Wilhelm next:', flip(*A), flip(*B))
print('PASS:', A[0]==B[0] and flip(*A)!=flip(*B))
```

```
mono_A={'sheet':0,'kappa':0,'history':''}
mono_B={'sheet':0,'kappa':2,'history':'g g'}
def next_g(s): return {'sheet':1-s['sheet'],'kappa':s['kappa']+1,'history':(s['history']+' g').strip()}
print('Monodromy projection:', mono_A['sheet'], mono_B['sheet'])
print('Monodromy next:', next_g(mono_A), next_g(mono_B))
print('PASS:', mono_A['sheet']==mono_B['sheet'] and next_g(mono_A)!=next_g(mono_B))
```

```
pair={'plus':1,'minus':-1}; empty={'plus':0,'minus':0}
def scalar(s): return s['plus']+s['minus']
def retained(s): return s['plus']
print('Shadow scalar projection:', scalar(pair), scalar(empty))
print('Shadow positive readout:', retained(pair), retained(empty))
print('PASS:', scalar(pair)==scalar(empty) and retained(pair)!=retained(empty))
```

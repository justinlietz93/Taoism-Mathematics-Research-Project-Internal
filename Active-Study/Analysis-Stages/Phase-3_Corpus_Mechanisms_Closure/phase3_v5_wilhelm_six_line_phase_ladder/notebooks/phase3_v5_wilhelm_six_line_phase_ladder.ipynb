# Phase 3 v5 — Wilhelm Six-Line Phase Ladder
Finite transition audit and projection-loss witnesses.

```python
from itertools import product
def flip_line(binary,line):
    idx=6-line
    b=list(binary); b[idx]='1' if b[idx]=='0' else '0'; return ''.join(b)
def hamming(a,b): return sum(x!=y for x,y in zip(a,b))
states=[''.join(bits) for bits in product('01', repeat=6)]
edges=[(s,l,flip_line(s,l)) for s in states for l in range(1,7)]
print('directed',len(edges),'undirected',len({tuple(sorted((s,t))) for s,l,t in edges}))
print('PASS', len(edges)==384 and all(hamming(s,t)==1 for s,l,t in edges))
```

```python
x=('111111',1); y=('111111',6)
print('projection',x[0],y[0])
print('next',flip_line(*x),flip_line(*y))
print('PASS projection loss', x[0]==y[0] and flip_line(*x)!=flip_line(*y))
```

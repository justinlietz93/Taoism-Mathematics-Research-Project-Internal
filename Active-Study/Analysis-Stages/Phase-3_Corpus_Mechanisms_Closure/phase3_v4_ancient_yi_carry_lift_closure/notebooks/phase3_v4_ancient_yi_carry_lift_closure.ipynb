# Phase 3 v4 — Ancient Yi Carry/Lift Closure
Each executable cell reports a gate and a result.

```python
def value_lsd(ds):
    return sum(d*(8**i) for i,d in enumerate(ds))

def succ(ds):
    ds=list(ds); i=0
    while True:
        if i==len(ds):
            ds.append(1); return tuple(ds), True
        ds[i]+=1
        if ds[i]<=7:
            return tuple(ds), False
        ds[i]=0; i+=1

examples = {'10':1,'67':62,'77':63,'001':64,'677':510,'777':511,'0001':512,'7777':4095}
rows=[]
for s,expected in examples.items():
    ds=tuple(int(c) for c in s)
    rows.append((s,value_lsd(ds),expected,value_lsd(ds)==expected))
print('G1_LSD_EXAMPLES_PASS', all(r[3] for r in rows))
rows
```

```python
def display(ds): return ''.join(str(d) for d in ds)
completion=[]
for k in range(1,9):
    all7=tuple([7]*k)
    nxt,lift=succ(all7)
    completion.append((display(all7), display(nxt), lift, nxt==tuple([0]*k+[1])))
print('G2_COMPLETION_LIFT_PASS', all(r[3] for r in completion))
completion
```

```python
witnesses = [
    ('low_digit', '7', '77', display(succ((7,))[0]), display(succ((7,7))[0])),
    ('scalar_zero', '0', '00', display(succ((0,))[0]), display(succ((0,0))[0])),
]
print('G3_PROJECTION_LOSS_PASS', all(w[3] != w[4] for w in witnesses))
witnesses
```
